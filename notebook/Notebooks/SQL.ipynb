# Connect to FHIR SQL Builder

```python
import intersystems_iris as iris
```

## Create the dbapi connection

```python
# create the dbapi connection

host = "iris"
port = 1972
namespace = "IRISAPP"
user = "SuperUser"
password = "SYS"

conn = iris.connect(
                    hostname=host, 
                    port=port, 
                    namespace=namespace, 
                    username=user, 
                    password=password
                   )

# create a cursor
cur = conn.cursor()
```

## Select Mol

```python
sql = """
SELECT 
* from Mol
"""
cur.execute(sql)
cur.fetchall()
```
